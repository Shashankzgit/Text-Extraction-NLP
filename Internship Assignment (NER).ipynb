### Data Cleaning:
* Creating a function using List of keywords to Extract out the necessary sentences from the whole list of sentences.
* Here the fn is applied on the column named 'seprated' containig list of sentenses.

```python
def fn1(x):
    X1=['adjacent','connected.','next']  #Found these keywords to be relative for information.
    L=[]
    for j in range(len(x)-1):
        Z=x[j].split()
        for i in Z:
            if i in X1:
                L.append(x[j])
    return L
```

### Building the Model:
* Here we are implementing Named Entity Recognition (NER) with spaCy.
* A custom model was trained on various examples to recognise Labeled entity for room and link Respectively. 

```python
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
train=[("washroom1 and livingroom1 are connected",{'entities':[(0,9,'Room'),(14,25,'Link')]}),
       ("washroom1 and bedroom2 are connected",{'entities':[(0,9,'Room'),(14,22,'Link')]}),
       ("kitchen1 and livingroom1 are connected",{'entities':[(0,8,'Room'),(13,24,'Link')]}),
       ("washroom2 and livingroom1 are connected",{'entities':[(0,9,'Room'),(14,25,'Link')]}),
       ("washroom1 and bedroom1 are connected",{'entities':[(0,9,'Room'),(14,22,'Link')]}),
       ("washroom1 and bedroom2 are connected",{'entities':[(0,9,'Room'),(14,22,'Link')]}),
       ("closet1 and bedroom1 are connected",{'entities':[(0,7,'Room'),(12,20,'Link')]}),
       ("kitchen1 and bedroom1 are connected",{'entities':[(0,8,'Room'),(13,21,'Link')]}),
       ("livingroom1 is next to closet1",{'entities':[(0,11,'Room'),(23,30,'Link')]}),
       ("washroom1 is adjacent to kitchen1",{'entities':[(0,9,'Room'),(25,33,'Link')]}),
       ("wbedroom1 is next to bedroom2",{'entities':[(0,9,'Room'),(21,29,'Link')]}),
       ("closet1 is next to washroom1",{'entities':[(0,7,'Room'),(19,28,'Link')]}),
       ("washroom1 is adjacent to washroom3",{'entities':[(0,9,'Room'),(25,34,'Link')]}),
       ("washroom3 and kitchen1 are connected",{'entities':[(0,9,'Room'),(14,22,'Link')]}),
       ("kitchen1 is next to washroom2",{'entities':[(0,8,'Room'),(20,29,'Link')]}),
       ("bedroom1 is adjacent to washroom1",{'entities':[(0,8,'Room'),(24,33,'Link')]}),
       ("bedroom1 is next to balcony1",{'entities':[(0,8,'Room'),(20,28,'Link')]}),
       ("kitchen1 is next to washroom1",{'entities':[(0,8,'Room'),(20,29,'Link')]})
       ]
```

```python
ner=nlp.get_pipe('ner') #Fecthing out the 'ner' pipe 
```

```python
for _, annotations in train:
    for ent in annotations.get("entities"):
        ner.add_label(ent[2]) #updating the 'ner' existing labels with the created entities.
```

```python
dis_pipes=[pipe for pipe in nlp.pipe_names if pipe !='ner'] 
#Storing other pipelines other than 'ner' in order to disabel them futher.
```

### Traning the model:

```python
import random
from spacy.util import minibatch,compounding 
from pathlib import Path

with nlp.disable_pipes(*dis_pipes):  #Disabeling the other pipelines while traning.
    optimizer=nlp.resume_training()  
    
    for iteration in range(100):     
        
        random.shuffle(train)
        losses={}
        
        batches=minibatch(train,size=compounding(1.0,4.0,1.001))
        for batch in batches:
            text,annotation=zip(*batch)
            nlp.update(text,
                      annotation,
                      drop=0.5,
                      losses=losses,
                      sgd=optimizer)
            print("Losses",losses)
```

### Creating column 'Link'.
* Creating a function to run through the cleaned col. containg the sentences.
* The fn. will extract out the List of corresponding Links to Rooms from sentence.

```python
def fn3(x):
    Li=[]
    for j in x:
        doc=nlp(j)
        for i in doc.ents:
            if i.label_=='Link':
                Li.append(i.text)
    return Li
```

```python
X['Link']=X['cleaned'].apply(fn3) #Created new column 'Link' by applying fn on Cleaned column.
```
